**SEO Optimization By Google search keyword scraping**

Websites appearance order appearance on google search depend on meta tags with the attribute name value = keywords.

To improve the SEO of a website this tag has to have a high quality content

So, my project focuses around getting the best keywords to use for a website depending on the type of the website.

So, the dataset we are going to work with would be the key words of the top google search results for a generic keyword for the website type

As we are going to have a lot of external scraping we would create a function to limit the time of a process so the code is efficient and doesn't take a lot of time when a url server is not responding

<ul>For that we are going to use the following libraries:
    <li> threading </li>
    <li> _thread </li>
    <li> contextmanager so we use the function using 'with' </li>
</ul>

```python
from contextlib import contextmanager
import threading
import _thread

class TimeoutException(Exception):
    def __init__(self, msg=''):
        self.msg = msg

@contextmanager
def time_limit(seconds):
    timer = threading.Timer(seconds, lambda: _thread.interrupt_main())
    timer.start()
    try:
        yield
    except KeyboardInterrupt:
        raise TimeoutException("Timed out for operation")
    finally:
        # if the action ends in specified time, timer is canceled
        timer.cancel()





```

Now to get our dataset we would need first to get the urls of the google search results

There some websites that we don't want to appear in our search results as they won't really add value to our data for example Wikipedia usually don't have the meta tag we look for, and the links that reference a google page in the search page, so it makes sense to exclude them from our search results.
for that we are going to create a tuplet containing the domains we would like to exclude

We would use regex to help work with strings and get clean urls from the search results as they are usually presented in a messy way

for that would create a function cleanLink()


```python
import re

#     excluding the domains we don't want as part of the search results
excluded_domains = ('https://www.google.', 
                  'https://google.', 
                  'https://webcache.googleusercontent.', 
                  'http://webcache.googleusercontent.', 
                  'https://policies.google.',
                  'https://support.google.',
                  'https://maps.google.',
                 'https://en.wikipedia.org')

    
# function to clean a link we get from the gooe search page
def cleanLink(url):
#    getting a clean url from the result messy url
    clean = re.search('\/url\?q\=(.*)\&sa',url['href'])
#   checking if the website is one of the exluded domains
    if clean != None:
        if not clean.group(1).startswith(excluded_domains):
            return clean.group(1)
        
    return None
```

Now we could create a function scrape_google() that returns the urls of the search results. the function would take the generic search keyword and the number of results wanted as parameters

<ul>We will use</ul>
    <li> <b>requests</b>: to get the data of the google search page </li>
    <li> <b>beautifulSoup</b>:  to store it and work with it in a simpler and neater way </li>
    <li> <b>urllib</b>:  to help in managing urls </li>



```python
from bs4 import BeautifulSoup
import requests
import urllib

# function to get links from google search
def scrape_google(query, number_result):

    linksList=[]
    page_number=0
    
#     checking if reached the number of results needed or further search is necessary
    while len(linksList)<number_result:
        
#     getting the data of result search page
        googleSearchUrl ="https://www.google.com/search?q="
        query = urllib.parse.quote_plus(query)
#         making sure url error won't interrupt the process
        try:
            response  = requests.get(googleSearchUrl + query+ "&start="+str(page_number*10), timeout=10)
        except:
            page_number+=1
        else:
            page_number+=1
            data = response.text
            soup = BeautifulSoup(data, "html.parser")
            links= soup.find_all('a', href=True)

            for link in links:
                clean = cleanLink(link)
    #           limiting the search results to the specified number
                if (clean != None) & (len(linksList)<number_result):
    #                   making sure it is not the same previous website or another page of it
                        if len(linksList) == 0:
                            linksList.append(clean)
                            i=0
                        else:
                            if not urllib.parse.urlparse(clean).netloc==urllib.parse.urlparse(linksList[i]).netloc:
                                linksList.append(clean)
                                i+=1
        
    return linksList
```

Before we store the keywords we need to filter these words even more to have a better dataset, so we don't handle the same words but with different casing as different entities eg.('Egypt', 'egypt'), they should be handled as the same word
For that we are going to unify the casing of the key words capitalizing only the first letter.

Also for instances that has the same meaning but different syntax eg.('job', 'jobs') should also be treated as the same word, as search engines do
For that we would use WordNetLemmatizer class from nltk library, to lemmatize all our keywords

we would create a separate function to filter the keywords filter_keywords()

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

nltk.download('stopwords')
cachedStopWords = stopwords.words("english")
lemmatizer = WordNetLemmatizer()

def filter_keywords(keywords):

    if not keywords == None:
        try:
#     splitting the keywords so that each word is an item in a list
            keywords=re.split(r'\W+',keywords['content'])
#           Applying the filters on the keywords
#           We lower the casing first to get a better performance of the lemmatize function
#           The we capitalize the result keyword
            keywords=[lemmatizer.lemmatize(word.lower()).capitalize() for word in keywords if word not in cachedStopWords]
            return keywords
        except:
            return []
    else:
        return []
    
```

Now, as we have the links of the search results and keywords filter function ready we would go into each page and get the values in the meta tag with the atrribute name=keywords, apply the filters on them and return a list of the result keywords by the function get_keywords()

```python
# function to get the keywords of a page
def get_keywords(url):
    
    try:
        response  = requests.get(url, timeout=10)
    except:
        keywords = None
    else:
        data = response.text
        soup = BeautifulSoup(data, "html.parser")
        keywords = soup.find('meta', attrs={"name": "keywords"})
    
    return filter_keywords(keywords)
```

Now, we will create a function google_keywords() to create a dictionary assigning each URL of the google results to it's keywords, the function would take the generic search keyword and the number of results wanted as parameters

```python
def google_keywords(query, number_result):
#     getting the links of the search results
    resultLinks=scrape_google(query, number_result)
    all_keywords={}
    
#     A visual to the user to show that the process is running or finished
    print("Getting data", end ="")
#     assin each link to it's keywords
    for link in resultLinks:
        print(".", end ="")
        all_keywords[link]=get_keywords(link)
    
    print("Process done")
    return all_keywords
```

Now that we have created the functions to get the keywords from the search results, we can apply them to get the keywords from the top search results of our topic, we can store this data in a json file so that we are to able to extract the info we need later

For my project I will look for the keywords for a website for tourism in Egypt, I will get a number of 50 data entites as an example

```python
import json

a_file = open("keywords_data.json", "w")
keywords_data=google_keywords("Egypt tourism", 50)
json.dump(keywords_data, a_file)
a_file.close()
```

Now that we have our dataset ready in a json file, we could load it so start working with the data inside

```python
with open('keywords_data.json', "r") as json_file:
    keywords_data=json.load(json_file)
```

As we are only interested in the keywords of the sites not the urls, so we are going to extract them and combine them in a single list

```python
all_keywords=[]
for key in keywords_data.keys():
    all_keywords.extend(keywords_data[key])
```

Now that we have all the keywords in a single list we are going to visualize them depending on the frequency of each keyword so the keywords with higher frequency would appear in a large font and vicevera 

<ul>For that we will use:
    <li> <b>pandas</b>: to get the data in a specific format </li>
    <li> <b>nltk & FreqDist</b>:  to calculate the frequency of the keywords </li>
    <li> <b>WordCloud & matplotlib</b>:  to visualize that frequency </li>
</ul>
